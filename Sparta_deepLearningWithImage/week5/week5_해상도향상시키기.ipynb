## pkg 로드하기

````python
import cv2
from google.colab.patches import cv2_imshow
````

## 모델 불러오기

````python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
sr.readModel('models/EDSR_x3.pb')
sr.setModel('edsr', 3)
````

## 원본이미지 불러오기

````python
img = cv2.imread('imgs/06.jpg')

print('input')
cv2_imshow(img)
````

## cv2의 resize함수로 이미지 3배 확대하기

````python
resized_img = cv2.resize(img, dsize=None, fx=3, fy=3)

print('resized_img::made by just resize()')
cv2_imshow(resized_img)
````

## 모델을 이용하여 이미지 3배 확대하기

````python
result = sr.upsample(img)

print('result::made by model')
cv2_imshow(result)
````
