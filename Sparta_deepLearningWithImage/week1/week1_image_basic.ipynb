## 원본이미지 불러오기

```python
import cv2
from google.colab.patches import cv2_imshow

img = cv2.imread('01.jpg')

print(img)
print(img.shape)
```

## 이미지에 사각형,원 그리기

```python
cv2.rectangle(img, pt1=(259,89), pt2=(380,348), color=(255,0,0), thickness=2)
cv2.circle(img, center=(320,220), radius=100, color=(0,0,255),thickness=3)

print('img')
cv2_imshow(img)
```

## 이미지 자르기

```python
cropped_img = img[89:348, 259:380]
print('crop')
cv2_imshow(cropped_img)
```

## 이미지 크기 줄이기

```python
img_resized = cv2.resize(img, (512, 256))

print('resized')
cv2_imshow(img_resized)
```

#이미지 컬러 시스템 변경하기

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)#BGR -> RGB

print('result')
cv2_imshow(img_rgb)
```
